```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
aluguel_casa = pd.read_excel("/content/houses_rent.xlsx")
aluguel_casa.head(10)
```

```python
aluguel_casa = aluguel_casa.rename(columns = {"parking space":"parking_space",
                                              "rent amount": "rent_amount",
                                              "property tax": "property_tax",
                                              "fire insurance": "fire_insurance"}
                                   )
aluguel_casa.head()
aluguel_casa.dtypes
```

```python
aluguel_casa = aluguel_casa.astype({"rent_amount":"float",
                                    "property_tax":"float",
                                    "fire_insurance":"float",
                                    "total":"float"}

                                    )
aluguel_casa.info()
```

```python
aluguel_casa[["area","rooms","bathroom","parking spaces", "rent_amount","property_tax","fire_insurance","total"]].describe()
```

```python
aluguel_casa = aluguel_casa.query("total <=4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept")]

aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept") &((aluguel_casa["rooms"]==3)|(aluguel_casa["rooms"]==2))]

aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
agg_city_price = aluguel_casa.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
ax = sns.barplot(data=agg_city_price, x = "rooms", y = "total")
ax.set(xlabel = 'Nmero de Quartos', ylabel = "Mdia do preo - Euro ");
```

```python
ax = sns.barplot(data=agg_city_price, x = "bathroom", y = "total")
ax.set(xlabel = 'Nmero de Banheiros', ylabel = "Mdia do preo - Euro ");
```
